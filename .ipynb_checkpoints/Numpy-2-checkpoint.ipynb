##### Creating an array

```python
# arrays are homogenous ( cant handle multiple data types together )
# convert list/tuple to create an array 
```

```python
ar1 =  np.array([10,20,30,40,50])
ar1
```

```python
np.zeros([3,4]) # 1 parameter # shape 
# create an array with 3 rows and 4 columns with zeros
```

```python
np.random.randint(1,10) # generate a random integer value from 1 to 10 # 2 parameters
```

```python
np.random.randint(1,10,[3,4]) # generate a random integer matrix using 1 to 10 in 3 rows and 4 columns
```

```python
# help(np.random.randint)
```

```python
np.random.random([3,4]) # generate a random matrix with 3 rows and 4 columns with random values between 0 to 1
```

```python
# python is dynamically typed prog language # we need not define data type anywhere
# create table id REAL
                # name varchar(255)
# BIGINT/SMALLINT : 64/32/128...
# numbers: int / float
# strings: U (numpy)
#          object (pandas)
```

```python
type(ar1)
```

```python
ar1 = np.arange(1,25)
ar1
```

```python
ar1.ndim
```

```python
ar1.size
```

```python
# converting 1D to 2D
ar2 = ar1.reshape(6,4) 
ar2
```

```python
ar2.ndim
```

```python
ar1.reshape(6,4)
# product of rows * columns == ar1.size 
```

```python
# As ar1 consists of 24 elements these combinations are possible 
# 1,24
# 2,12
# 3,8
# 4,6
# ..
```

```python
ar1.reshape(5,5)
# 5*5 != 24
```

```python
ar1.reshape(3,3)
# 3*3 != 24
```

```python
ar3 = ar1.reshape(3,2,4) # create 3 sub matrices with 2 rows 4 columns
# 3*2*4 == 24
ar3
```

```python
ar3.ndim
```

```python
ar1.reshape(2,3,4)
```

```python
ar1 = np.arange(1,25)
ar1.reshape(6,4)
```

```python
# better way
ar1 = np.arange(1,25).reshape(6,4)
ar1
```

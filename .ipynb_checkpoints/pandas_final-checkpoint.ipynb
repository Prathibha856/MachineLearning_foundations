```python
# Create a DataFrame of employees
# Name , dept, salary
# Display all the employees  working in the IT department
# add a new column Bonus = 10 % of salary
# Find the average salary of the employees
# Change the column name from Salary to CTC.
```

```python
data = {
    'name': ['Vansh', 'Dhruv', 'Virat', 'Rohit', 'Dhoni'],
    'dept': ['HR', 'IT', 'Finance', 'IT', 'Marketing'],
    'salary': [50000, 60000, 55000, 65000, 52000]
}
df = pd.DataFrame(data)
it_employees = df[df['dept'] == 'IT']
print("Employees in IT department:")
print(it_employees)
df['bonus'] = df['salary'] * 0.10
average_salary = df['salary'].mean()
print(f"\nAverage salary of employees: {average_salary}")
df.rename(columns={'salary': 'ctc'}, inplace=True)
print("\nUpdated DataFrame:")
print(df)
```

```python
# Create a Data frame with columns sports and Age
# Take atleast 5 values
# search/filter for the name/names where the age is 20
# sort the dataframe in both ascending and decending order of age
# using python loops, print the ages from the dataframe and store in a set
```

```python
df = pd.DataFrame({
    "sport": ["cricket", "football", "badminton", "basketball", "volleyball"],
    "age": [19, 22, 20, 50, 42]
})
print(df)
print(df[df['age'] == 20])
print(df.sort_values(by = "age"))
print(df.sort_values(by = "age",ascending=False))
sets = set()

for age in df["age"]:
    sets.add(age)

print(sets)
```

```python
# Create two separate data Frames A and B
# concatenate the 2 dataframes to a new dataframe C
# Split the dataframe  C row wise
```

```python
import pandas as pd

A = pd.DataFrame({
    "Name": ["Asha", "Rahul", "Sneha"],
    "Age": [21, 22, 20]
})

A
```

```python
B = pd.DataFrame({
    "Name": ["Kiran", "Meera", "Arjun"],
    "Age": [23, 24, 22]
})

B
```

```python
C = pd.concat([A, B], ignore_index=True)

C
```

```python
#Why ignore_index=True?
#Because otherwise indexes would repeat:

#0 1 2
#0 1 2


#With ignore_index=True  new clean index:

#0 1 2 3 4 5
```

```python
# Split row-wise
mid = len(C) // 2
C1 = C.iloc[:mid]
C2 = C.iloc[mid:]
```

```python
print(C1)
```

```python
print(C2)
```

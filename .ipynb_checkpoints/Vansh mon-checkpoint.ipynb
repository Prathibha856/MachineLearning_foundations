```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/cars (1) (2).csv',skipinitialspace=True)
```

```python
#Dataset link -> https://drive.google.com/file/d/18PV-dMn2y1KYCdZ0c9JoOnvEh4MIxeGy/view?usp=sharing
```

```python
df
```

```python
# Create a boxplot
# x - cylinders
# y - hp
# hue - brand
# also find the unique values in the brand column
```

```python
sns.boxplot(data = df,x = "cylinders",y = "hp", hue="brand")
plt.title("Boxplot of Cylinder and HP")
plt.show()
```

```python
df['brand'].unique()
```

```python
# Create a heatmap for the entire data to study the correlation between each of the columns
# Time till 9:57 pm
# Heatmap only works with numerical data
```

```python
df.info()
```

```python
numData = df.select_dtypes(include=[float,int])
corrMat = numData.corr()
corrMat
```

```python
sns.heatmap(corrMat)
plt.show()
```

```python
sns.heatmap(corrMat,annot=True,cmap="Blues")
plt.show()
```

```python
# Create a pie plot for Brand columns with the total values of their freq(value_counts).
# try to use expode
# autopct
# time till 10:20 pm
```

```python
x = df['brand'].value_counts()
```

```python
x
```

```python
x.plot.pie(autopct = '%.2f%%',explode=[0.2,0.2,0.2],shadow=True)
plt.show()
```

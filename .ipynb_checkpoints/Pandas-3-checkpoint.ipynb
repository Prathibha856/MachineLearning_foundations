```python
def sqr_num(num):
    return num**2
```

```python
df1['acceleration'].apply(sqr_num)
```

```python
def evn_odd(num):
    if num%2 ==0:
        return 'Even'
    else:
        return 'Odd'
```

```python
df1['col1'] =  df1['acceleration'].apply(evn_odd)
df1.head(2)
```

```python
df1['col2'] =  df1['acceleration'].apply(lambda num: num**2)
df1['col3'] =  df1['acceleration'].apply(lambda num: 'Even' if num%2 ==0 else 'Odd')
df1.head()
```

```python
uniq = [1,2,3]
count = [248,70,79]
```

```python
plt.bar(uniq, count,color = ['Red','Blue','Black'])
plt.xlabel('Origin')
plt.ylabel('Freq')
plt.title('Bar Plot')
plt.show()
```

```python
plt.barh(uniq, count,color = ['Red','Blue','Black'])
plt.xlabel('Origin')
plt.ylabel('Freq')
plt.title('Bar Plot')
plt.show()
```

```python
output = df1['origin'].value_counts() # output is series format
output
```

```python
type(output)
```

```python
uniq = output.index
```

```python
uniq
```

```python
freq = output.values
freq
```

```python
plt.bar(uniq , freq)
```

```python
##### PIE CHArt - used for the discrete datas
```

```python
plt.pie(freq , labels = uniq)
plt.show()
```

```python
# only bar , pie is for discrete and others all continuos only
```

```python
plt.pie(freq , labels = uniq , autopct= '%.2f') # 2 float/decimals
plt.show()
```

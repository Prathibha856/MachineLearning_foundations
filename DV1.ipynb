```python
import matplotlib.pyplot as plt
```

```python
A= [1,2,3,4,5,6]
B= [1,2,5,6,4,9]
C=[1,6,8,6,9,7]
```

```python
plt.scatter(C,A)
```

```python
plt.scatter(A,B)
```

```python
plt.scatter(B,A , color='Red')
plt.xlabel('Sales of manager')
plt.ylabel('salary of that person')
plt.title('ALL about sales')
plt.show()
```

```python
plt.scatter(B,A , color='Red' ,marker ='D' , label ='Person A')
plt.scatter(A,B, color='green',marker = 'o', label= 'Person B')
plt.xlabel('Sales of manager')
plt.ylabel('salary of that person')
plt.title('ALL about sales')
plt.show()
```

```python
plt.scatter(B,A , color='Red' ,marker ='D' , label ='Person A') # when labeled not mentioneed legend gets error
plt.scatter(A,B, color='green',marker = 'o', label= 'Person B')
plt.xlabel('Sales of manager')
plt.ylabel('salary of that person')
plt.title('ALL about sales')
plt.grid()
plt.legend()# displayes values entered in labels --> legends are required
plt.show()
```

```python
plt.subplot(221)
plt.scatter(B,A , color='Red' ,marker ='D' , label ='Person A') # when labeled not mentioneed legend gets error

plt.xlabel('Sales of manager')
plt.ylabel('salary of that person')
plt.title('ALL about sales')
plt.grid()
plt.legend()# displayes values entered in labels --> legends are required


plt.subplot(222) # number of rows, columns and then graph numbers
plt.plot(A,B, color='green',marker = 'o', label= 'Person B')
plt.xlabel('Sales of manager')
plt.ylabel('salary of that person')
plt.title('ALL about sales')
plt.grid()
plt.legend()# displayes values entered in labels --> legends are required


plt.subplot(223)
plt.plot(B,A , color='Red' ,marker ='D' , label ='Person A') # when labeled not mentioneed legend gets error

plt.xlabel('Sales of manager')
plt.ylabel('salary of that person')
plt.title('ALL about sales')
plt.grid()
plt.legend()# displayes values entered in labels --> legends are required


plt.subplot(224)
plt.stackplot(A,B, color='green')
plt.xlabel('Sales of manager')
plt.ylabel('salary of that person')
plt.title('ALL about sales')
plt.grid()
# legend shouldnot be there in the stackplot plt.legend()# displayes values entered in labels --> legends are required

plt.show()
```

```python
fig , axs = plt.subplots(2,2)
#fig: is used to save/export the graph
#axs: is used to identify index
```

```python
fig , axs = plt.subplots(2,2)
axs[0,0].scatter(B,A)
axs[0,1].plot(B,A)
axs[1,0].plot(A,B)
axs[1,1].stackplot(A,B)
plt.subplots_adjust(wspace=0.8 , hspace=0.6)
plt.grid()
plt.show()
```
